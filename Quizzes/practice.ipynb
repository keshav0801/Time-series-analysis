```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from statsmodels.graphics.tsaplots import plot_acf
from statsmodels.graphics.tsaplots import plot_pacf

plt.style.use(['seaborn'])
```

```python
et = np.random.normal(0,1,size=1000)
yt = []
yt.append(2 + et[0])
for t in range(1,100):
    yt.append(2 + 0.3*t + et[t])
plt.plot(yt)
plt.show()
```

```python
ytnew =[]
ytnew.append(et[0])
for t in range(1,100):
    ytnew.append(1*ytnew[t-1] +et[t])
for t in range(100,200):
    ytnew.append(-1*ytnew[t-1] + et[t])
plt.plot(ytnew)
plt.show()
```

```python
ytnew1 = np.array(ytnew[100:])
np.mean((ytnew1[2:]-np.mean(ytnew1))*(ytnew1[:-2]-np.mean(ytnew1)))
```

```python
plot_pacf(ytnew)
plt.show()
```

```python
plot_pacf(ytnew[:100])
plt.show()
```
