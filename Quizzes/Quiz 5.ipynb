```python
e = np.random.normal(0,2,size=500)
time = np.arange(500)
```

```python
yt_pos = [] # Positive autocorrelation time series plot
yt_pos.append(e[0])
for i in range(1,500):
    yt_pos.append(yt_pos[i-1] + e[i])
```

```python
plt.scatter(time,yt_pos)
```

It can be seen that the positive autocorrelation is a random walk process, where yt moves in the same direction as the previous term, but is either moving up or down depending on the error term added in each subsequent term. Autocorrelation function is almost perfectly 1. Yt is being pushed closer to the previous term.

```python
yt_neg = [] # negative autocorrelation time series plot
yt_neg.append(e[0])
for i in range(1,500):
    yt_neg.append(-yt_neg[i-1] + e[i])
```

```python
plt.scatter(time,yt_neg)
```

Here each term is a negative of the previous term, hence we get a mirror like graph of the previous function. Here the function is oscillating on each term as each term is the negative of previous one plus the error term pushing it either up or down.
Here, each term is pushed away from the previous term hence we get a shape like this. Autocorrelation of order 1 is negative in this graph.
